```python
import numpy as np
from numpy import asarray
from PIL import Image
from scipy import stats
import webcolors
import pandas as pd
from operator import itemgetter
import statistics
```

```python
def closest_colour(requested_colour):
    min_colours = {}
    for key, name in webcolors.CSS3_HEX_TO_NAMES.items():
        r_c, g_c, b_c = webcolors.hex_to_rgb(key)
        rd = (r_c - requested_colour[0]) ** 2
        gd = (g_c - requested_colour[1]) ** 2
        bd = (b_c - requested_colour[2]) ** 2
        min_colours[(rd + gd + bd)] = name
    return min_colours[min(min_colours.keys())]


def get_colour_name(requested_colour):
    try:
        closest_name = actual_name = webcolors.rgb_to_name(requested_colour)
    except ValueError:
        closest_name = closest_colour(requested_colour)
        actual_name = None
    return closest_name


def get_bbox_color(x1, y1, x2, y2, image_path):
    image = Image.open(image_path)
    width, height = image.size
    pixel_data = asarray(image)


    bbox = pixel_data[(height-y2):(height-y1), x1:x2]
    bbox_mode = stats.mode(bbox, keepdims=False)
    bbox_mode = tuple(bbox_mode[0][0])

    return get_colour_name(bbox_mode)
```
