## Clip Counter
Counts the number of clips per class

```python
count_path = 'var_annot_jsons/17-03-2022-16_23_40_master.json'
```

```python
#program to count the label totals for each individual annotation (cr: 37, ar: 21, ru: 34, etc.)

df = pd.read_json(count_path, orient = 'index')


def dict_2_str(dicc):
    dicc = str(dicc)
    dicc = dicc.split(',')
    return dicc[0]

df['annotations'] = df['annotations'].apply(dict_2_str)

print(df['annotations'].value_counts())
print('total annotations: ', df['annotations'].value_counts(). sum())
print(df['annotations'].value_counts(normalize = True))
```

```python
json_path = 'var_jsons/17-03-2022-16_23_40.json'
```

```python
# Read json and clean data

df = pd.read_json(json_path, orient = 'index')
df = df.reset_index()
df = df.rename(columns={'index': 'youtube_id', 'subset': 'split'})
```

```python
#Average length of clip

total_time = df['duration'].sum()
avg_time = total_time/len(df)
print('average clip length: ', round(avg_time,2))
```

```python
# Clip length distribution

dur = []
dur2 = []
dur5 = []
dur10 = []
dur50 = []

for i in range(0, len(df)):
    annot = df.iloc[i]['annotations'], df.iloc[i]['duration']
    duration = annot[1]
    dur.append(duration)
    if duration <= 2:
        dur2.append(duration)
    elif duration <= 5: 
        dur5.append(duration)
    elif duration <= 10:
        dur10.append(duration)
    else:
        dur50.append(duration)

perc2 = round(len(dur2)/len(dur),3)
perc5 = round(len(dur5)/len(dur),3)
perc10 = round(len(dur10)/len(dur),3)
perc50 = round(len(dur50)/len(dur),3)

print('clip length distribution:')
print('0-2s:  ', perc2)
print('2-5s:  ', perc5)
print('5-10s: ', perc10)
print('10+:   ', perc50) 

# Pie chart

labels = '0-2s', '2-5s', '5-10s', '10+'
sizes = [perc2, perc5, perc10, perc50]
explode = (0, 0.1, 0, 0)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')
plt.title('Clip Length Distribution')

plt.show()
```

```python
# Total duration per class

total_lst = []
crowd_lst = []
arrest_lst = []
brawl_lst = []
pog_lst = []
strike_lst = []
throw_lst = []
spray_lst = []
aim_lst = []

for i in range(0, len(df)):
    annot = df.iloc[i]['annotations'], df.iloc[i]['duration']
    label = annot[0]['label']
    duration = annot[1]
    total_lst.append(duration)
    if label == 'crowd':
        crowd_lst.append(duration)
    elif label == 'restraining':
        arrest_lst.append(duration)
    elif label == 'brawling':
        brawl_lst.append(duration)
    elif label == 'person_on_ground':
        pog_lst.append(duration)
    elif label == 'striking':
        strike_lst.append(duration)
    elif label == 'throwing':
        throw_lst.append(duration)
    elif label == 'spray':
        spray_lst.append(duration)
    elif label == 'aiming':
        aim_lst.append(duration)
    
cr_sum = round(sum(crowd_lst)/(60),2)
ar_sum = round(sum(arrest_lst)/(60),2) 
br_sum = round(sum(brawl_lst)/(60),2)
pg_sum = round(sum(pog_lst)/(60),2)
st_sum = round(sum(strike_lst)/(60),2)
th_sum = round(sum(throw_lst)/(60),2)
ps_sum = round(sum(spray_lst)/(60),2)
am_sum = round(sum(aim_lst)/(60),2)

print('total duration per class:')
print('crowd sum:            ', cr_sum, 'min')
print('arresting sum:        ', ar_sum, 'min')
print('brawling sum:         ', br_sum, 'min')
print('person on ground sum: ', pg_sum, 'min')
print('striking sum:         ', st_sum, 'min')
print('throwing sum:         ', th_sum, 'min')
print('spray sum:            ', ps_sum, 'min')
print('aiming sum:           ', am_sum, 'min')
print('total sum:            ', round(sum(total_lst)/(60),2), 'min') 

#Bar chart

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
classes = ['cr', 'ar', 'br', 'pg', 'st', 'th', 'ps', 'am']
totals = [cr_sum, ar_sum, br_sum, pg_sum, st_sum, th_sum, ps_sum, am_sum]
ax.bar(classes, totals)
plt.title('Total Duration per Class', color = 'black')
plt.ylabel('Minutes')
plt.xlabel('Class')
plt.show()
```

```python
# Average clip length by class

avg_cr = round(sum(crowd_lst)/len(crowd_lst),2)
avg_ar = round(sum(arrest_lst)/len(arrest_lst),2)
avg_br = round(sum(brawl_lst)/len(brawl_lst),2)
avg_pg = round(sum(pog_lst)/len(pog_lst),2)
avg_st = round(sum(strike_lst)/len(strike_lst),2)
avg_th = round(sum(throw_lst)/len(throw_lst),2)
avg_ps = round(sum(spray_lst)/len(spray_lst),2)
avg_am = round(sum(aim_lst)/len(aim_lst),2)

print('average clip length by class:')
print('crowd:            ', avg_cr)
print('arresting:        ', avg_ar)
print('brawling:         ', avg_br)
print('person on ground: ', avg_pg)
print('striking:         ', avg_st)
print('throwing:         ', avg_th)
print('spray:            ', avg_ps)
print('aiming:           ', avg_am)

#Bar chart

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
classes = ['cr', 'ar', 'br', 'pg', 'st', 'th', 'ps', 'am']
totals = [avg_cr, avg_ar, avg_br, avg_pg, avg_st, avg_th, avg_ps, avg_am]
ax.bar(classes, totals)
plt.title('Average Clip Length by Class', color='black')
plt.ylabel('Seconds')
plt.xlabel('Class')
plt.show()
```
