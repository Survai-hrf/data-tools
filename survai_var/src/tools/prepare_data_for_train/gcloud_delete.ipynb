```python
import os
import pandas as pd
from google.cloud import storage
```

```python
csv_path = 'var_data.csv'

df = pd.read_csv(csv_path)
df = df.reset_index()
df['clarity_level'] = df['clarity_level'].fillna('none') 
df['split'] = 'train'
df.head()
```

```python
os.environ['GOOGLE_APPLICATION_CREDENTIALS'] = 'survai-data-connect.json'
storage_client = storage.Client()
bucket_name = 'survai-dataset'
bucket = storage_client.bucket(bucket_name)
```

```python
videos = storage_client.list_blobs('survai-dataset', prefix='datasets/var/master_videos/')
video_list = [video.name for video in videos]
video_list
```

```python
for ind, row in df.iterrows():
    id = row['id']
    label = str(row['label'])
    start = int(row['time_start'])
    end = int(row['time_end'])
    fill_start = str(start).zfill(6)
    fill_end = str(end).zfill(6)

    if id[0] == '-':
        file_name = f'{id[1:]}_{fill_start}_{fill_end}.mp4'
        bucket = storage_client.bucket(bucket_name)
        blob = bucket.blob(f"datasets/var/master_videos/{label}/{file_name}")

        if f'datasets/var/master_videos/{label}/{file_name}' in video_list:
            blob.delete()
            print(file_name, 'has been CANNED ;)')
```
