```python
import pandas as pd
import json
```

```python
# load data
json_path = 'jsons/ground_truth.json'

with open(json_path) as d:
    data = json.load(d)

data['annotations']
```

```python
categories = {
    0: 'uniformed',
    1: 'non_uniformed',
    2: 'riot_shield',
    3: 'gun',
    4: 'pepper_spray',
    5: 'baton',
    6: 'chemical_smoke'
}
```

```python
# create containers to store reformatted data
csv_store = []

for annot in data['annotations']:
    csv_store.append({
        'label_name': categories.get(annot['category_id']),
        'bbox_x1': annot['bbox'][0],
        'bbox_y1': annot['bbox'][1],
        'bbox_x2': annot['bbox'][2],
        'bbox_y2': annot['bbox'][3],
        'image_name': data['images'][annot['image_id']]['file_name'],
        'width': data['images'][annot['image_id']]['width'],
        'height': data['images'][annot['image_id']]['height']
    })
```

```python
df = pd.DataFrame(csv_store)
df.head(20)
```

```python
df.to_csv('ground_truth.csv')
```
